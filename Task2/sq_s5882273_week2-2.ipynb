## 2.12

```python
from math import atan2, degrees, pi
x_val = input('Enter x:')
x = float(x_val)
y_val = input('Enter y:')
y = float(y_val)

angle = atan2(y,x)
angle_deg = degrees(angle)
print(f'The angle from the x-axis is: {angle_deg:5.3f} degrees')

# difference between atan and atan2: atan gives the arctan of whatever value you put in it. atan(x)
# atan2 is the arctan of the first argument over the second argument. atan(y,x)

if angle>=pi/2:
    quad="second"
    
elif angle>=0:
    quad="first"
        
elif angle<=-pi/2:
    quad="third"
    
else :
    quad="fourth"
    
print("It is located in the " +quad+ " quadrant")
```
